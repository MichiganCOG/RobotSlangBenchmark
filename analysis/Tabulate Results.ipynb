```python
import glob
import pandas as pd
csvs = glob.glob('csvs/*')
data = [pd.read_csv(c) for c in csvs]
```

```python
import numpy as np
def remove_nans(arr):
    new_arr = np.zeros(4)
    rows, cols = arr.shape
    for i in range(rows):
        for j in range(cols):
            if not np.isnan(arr[i,j]):
                new_arr[j] = arr[i,j]
    return new_arr.tolist()
```

```python
# Make a new dataframe that removes NANs
from collections import defaultdict 


new_data = defaultdict(list)
keys = ['Name', 'val_seen_top_final_goal', 'val_seen_top_oracle_goal', 'test_top_final_goal', 'test_top_oracle_goal']
for c,df in zip(csvs, data):
    c = c.split('/')[1].replace('.csv', '')
    df = df[keys]
    for index, row in df.iterrows():
        name, *values = row.values.tolist()
        name = c + '--' + name.replace('test-','').replace('val-', '')
        new_data[name].append(values)

# Convert to arrays
new_data = [[k] + remove_nans(np.asarray(v)) for k,v in new_data.items()]
new_data = pd.DataFrame(data=new_data, columns=keys)
```
